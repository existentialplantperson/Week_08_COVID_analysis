```python
artists = ["coldplay", "prince", "the%20rolling%20 stones", "drake", "aerosmith", "radiohead", "neko%20case", "adele",
           "beyonce", "mercyme", "madonna" "pink%20floyd", "ac/cd", "foo%20fighters", "linkin%20park", "daft%20punk", 
           "queen", "michael%20jackson", "metallica", "the%20prodigy", "muse", "david%20bowie", "red%20hot%20chili%20peppers",
           "2pac", "rihanna", "gorillaz", "u2", "the%20xx", "abba", "katy%20perry", "the%20beatles", "eminem", "imagine%20dragons"
          "nightwish", "evanescence","pearl%20jam", "bruno%20mars", "lindsey%20striling", "faithless", "rammstein", "harry%20styles", "death", "nirvana", "halsey"
          "sam%20smith", "meek%20", "normani", "kid%20cudi", "billie%20eilish", "lauren%20daigle", "the%20weeknd", "khalid", 
           "taylor%20swift", "josh%20groban", "sting", "lady%20gaga", "maya", "fugees", "lauryn%20hill", "marvin%20gaye", "amy%20winehouse" ]
```

```python
audio_data = {'artist id':[],
              'artist':[],
              'gender':[],
              'members':[],
              'style':[], 
              'genre':[],
             'year_formed':[],
             'year_disbanded':[],
             'location':[]}

#loop over artists dictionary and add data to dictionary
for artist in artists:
    base_url = "http://theaudiodb.com/api/v1/json/1/search.php?s="

    try:
        req = requests.get(base_url + f'{artist}')
        data = req.json()
        #url = urllib3.urlopen(base_url + f'{artist}')
        #data = json.load(url)
        
        for item in data['artists']:
        #add values to dictionary
            #print(["strArtist"])
            #print(["strStyle"])
            #print(["strGenre"])
        
            audio_data['artist id'].append(item["idArtist"])
            audio_data['artist'].append(item["strArtist"])
            audio_data['gender'].append(item["strGender"])
            audio_data['members'].append(item["intMembers"])
            audio_data['style'].append(item["strStyle"])
            audio_data['genre'].append(item["strGenre"])
            audio_data['year_formed'].append(item["intFormedYear"])
            audio_data['year_disbanded'].append(item["strDisbanded"])
            audio_data['location'].append(item["strCountry"])
        
            
            #audio_data['artist'].append(item["artists"]["strArtist"])
            #audio_data['style'].append(item["artists"]["strStyle"])
            #audio_data['genre'].append(item["artists"]["strGenre"])
        
        print(base_url + f'{artist}')
        
    except:
        pass
```
